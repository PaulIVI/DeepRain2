```python
import os
import json
```

```python
Model1 = "30min_LSTM_znBinomial"
Model2 = "zeroInf_negative_Binomial"
pwd = "Models_weights"
```

```python
first = os.path.join(pwd,Model1)
second = os.path.join(pwd,Model2)
```

```python
hist_first = os.path.join(first,Model1+"_history.json")
hist_second = os.path.join(second,Model2+"_history.json")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.pyplot import figure
axis_font = {'size':'25'}
```

```python
def plotAUC(baseline_dict,labels,title,lw=4):
    from sklearn.metrics import roc_curve, auc
    
    figure(num=None, figsize=(20, 15), dpi=100, facecolor='w', edgecolor='k')

    sns.set(style="ticks", context="talk")
    sns.set_style("darkgrid")
    plt.rcParams.update(params)
    for i,b in enumerate(baseline_dict):
        
        tp = []
        fp = []
        
        for key in b:
            base = b[key]
            tp.append(base["TP"]/(base["TP"]+base["FN"]) )
            fp.append(base["FP"]/(base["FP"]+base["FN"]))
    
        plt.plot(fp, tp,lw=lw, label='{} (area = {:.2f})'.format(labels[i],auc(fp,tp)))
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC - '+title)
    plt.legend(loc="lower right")
    plt.show()


plotAUC([unet_baseline,LSTM_baseline],["Unet ROC","LSTM ROC"],title="Zero-Inflated negative Binomial" )
```
